### Basic Plot

[Top](#Section)

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.show()
#Notice the x axis goes from 0 to 3, and the y from 0 to 3. This is because matplotlib takes y as the sequence of
#value and auto generate the x axis. Python range starts with 0 and has same length as y, hence [0,1,2,3]
```

```python
#How it would be done using OO
fig, ax = plt.subplots()
ax.plot([1,2,3,4])
ax.set_ylabel('some numbers')
```

```python
#plot can take any number of argument and graph them
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

### Plotting With Keyword
When you data you can access by keyword, you can provide the object with the "data" keyword to generate the plot

[Top](#Section)

```python
data = {'a': np.arange(5),
        'c': np.random.randint(0, 10, 5),
        'd': np.random.randn(5)}
data['b'] = data['a'] + 10 * np.random.randn(5)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data) #If you look at the pyplot doc, c stands for color (should be int), and s for size
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

### Plotting Catagorical Variables

[Top](#Section)

```python
month = ['Jan', 'Feb', 'Mar','Aprl', 'May', 'Jun' ]
balance = [100,95,80,75,68,92]

plt.figure(figsize=(10,3))
plt.subplot(131) #one row, three column, set first column to
plt.bar(month,balance)
plt.subplot(132)
plt.scatter(month,balance)
plt.subplot(133)
plt.plot(month,balance,'s')
plt.suptitle('Categorical Plotting')
plt.show
```

### Multiple Figures and Axes

[Top](#Section)

```python
balance1 = np.random.randint(75,150,12)
balance2 = np.random.randint(75,150,12)
plt.figure()
plt.subplot(211)
plt.plot(range(1,13), balance1, 'bo', range(1,13), balance2,'k')
plt.subplot(212)
plt.plot(range(1,13), np.arange(80,150,6))
```

```python
#You Can Also Set Multiple Figures
a = np.arange(1,20,.5)
plt.figure(1) #first fig
plt.subplot(211)
plt.plot(a,a,'g-')

plt.figure(2) #second fig
plt.plot(a,a**2, 'rx')
plt.grid(True)

plt.figure(1) #calls back the first fig
plt.subplot(211) #set subplot fig1 to current fig
plt.title('Linear') #set title for that fig
```

## Annotating Text

[Top](#Section)

```python
month = ['Jan', 'Feb', 'Mar','Aprl', 'May', 'Jun' ]
balance = [100,95,80,75,68,92]

plt.figure(figsize=(10,3))
plt.plot(month,balance, lw=3)
plt.suptitle('Using Annotation')
plt.annotate('Bottom', xy=("May", 68), xytext=(4.7,68), arrowprops=dict(arrowstyle="->"))
#plt.annotate('text', xy = point to annotate), xytext = position to place text, fancy arrow properties)
plt.show
```
