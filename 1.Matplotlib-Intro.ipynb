## Simpliest way to create a plot

The simpliest way to crate a plot using Matplotlib is to create a figure and an axes. Figure can be thought as the graph with the x and y axis, and the axes is the area where you can plot your data using the x and y corodinates.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fig, ax = plt.subplots()
```

```python
#As you can see we have the figure with the axis, but we havent declares any axes to plot.
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[22,30,35,29])
```

```python
#You can have more than one axes
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[4,8,15,16])
ax.plot([4,5,6,7],[16,23,42,22])
```

## Ways to Plot

There are two ways to plot:<br>
<ol>
    <li>Object oriented way where we create figure, axes and call methods on them<br>
         fig, ax = plt.subplots()<br>
         ax.plot([1,2,3,4],[4,8,15,16], label='OO-way')
<br>        
    <li>Use pyplot to automatically create and manage the figures and axes using pyplot functions for plotting<br>
        plt.plot([1,2,3,4],[4,8,15,16], label='pyplot way')

### 1) Object Oriented way

```python
x = [1,2,3,4]

fig, ax = plt.subplots()  
ax.plot(x, x, label='linear')  
ax.plot(x, [(y+1)**2 for y in range(4)], label='quadratic') 
ax.plot(x, [(y+1)**3 for y in range(4)], label='cubic') 
ax.set_xlabel('x label')  #Add x-label 
ax.set_ylabel('y label')  #Add y-label
ax.set_title("OO-Simple Plot")  #Add a title 
ax.legend()  # Add legend.
```

### 2)Using pyplot

```python
plt.plot(x, x, label='linear')  
plt.plot(x, [(y+1)**2 for y in range(4)], label='quadratic') 
plt.plot(x, [(y+1)**3 for y in range(4)], label='cubic') 
plt.xlabel('x label')  #Add x-label 
plt.ylabel('y label')  #Add y-label
plt.title("plt-Simple Plot")  #Add a title 
plt.legend()  # Add legend
plt.show()
```

## Functions with plots

```python
#1) Create a function that takes two data sets and plot them on an axes
def plotter(ax, data1, data2, param_dict):
    to_plot = ax.plot(data1, data2, **param_dict) #the ** param_dict is like on the variables that will be 
                                                  #assign in this postion. Here we will pass a dictionary 
                                                  #to this parameter potion
    return to_plot

#2 Create the data
data1, data2 = np.random.randn(2, 100) #return 2 random list with 100 entrys

#3 Create the figure and axes
fig, ax = plt.subplots(1,1) #In the example above we used the defualt subplot, here we are setting a 1x1 subplot

#4 Call the function
plotter(ax, data1, data2, {'marker': 'x'})
```

```python
#Add another figure
fig, (ax1, ax2) = plt.subplots(1, 2)
data3,data4 = np.random.randn(2,100)
plotter(ax1, data1, data2, {'marker': 'x'})
plotter(ax2, data3, data4, {'marker': 'o'})
```
