## OO API vs Pyplot

[Top](#Section)

Remember that there are two ways to plot your data. The first is using the object oriented interface which calls an instatnce of axes.Axes to render visualization on instance of figure.Figure. The second is using the pyplot module, state-based interface. In gerneral the oo interface is perferred in the tutorial and I personally like it as well.

## The Data
This will be copied from the website as to ensure we use the same data

[Top](#Section)

```python
import numpy as np
import matplotlib.pyplot as plt


data = {'Barton LLC': 109438.50,
        'Frami, Hills and Schmidt': 103569.59,
        'Fritsch, Russel and Anderson': 112214.71,
        'Jerde-Hilpert': 112591.43,
        'Keeling LLC': 100934.30,
        'Koepp Ltd': 103660.54,
        'Kulas Inc': 137351.96,
        'Trantow-Barrows': 123381.38,
        'White-Trantow': 135841.99,
        'Will LLC': 104437.60}
group_data = list(data.values())
group_names = list(data.keys())
group_mean = np.mean(group_data)
```

### Canvase
Before we can start ploting, we need to call the axes instance to write to the figure.Figure instance. In a way we are using ax to draw on the firgure, think of this as a blank canvase.

[Top](#Section)

```python
fig, ax = plt.subplots()
```

## Painting
Now that you have your canvase, we next need to draw the data with ax.

[Top](#Section)

```python
fig, ax = plt.subplots()
ax.barh(group_names, group_data) 
#Here we use the ax instance to draw a, bar chart that is horizontal, with name on the y axis and data on the x axis
```

## Style
"The style controls many things, such as color, linewidths, backgrounds, etc."

[Top](#Section)

```python
#To see a list of available styles
print(plt.style.available)
```

```python
plt.style.use('bmh')
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
```

## Customize the Plot-Axis
To customize the axis, first we want to get the lables of say the x tick marks using the get_xticklabels() function

[Top](#Section)

```python
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels() #Here we are getting the xticklables and storing them in lables
```

## X-axis Rotation
We can utilize the pyplot.setp() funciton. This takes one or many list of Maplotlib objects and tries to set the style for each one.

[Top](#Section)

```python
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels() #Here we are getting the xticklables and storing them in lables
plt.setp(labels, rotation=45, horizontalalignment='right')
#Using plt.setp, we pass the lables of x, and rotate them 45 degrees and align them to the right
```

## Chart Layout

If we had used andother style it may have cut off the some of the chart. To remedy this, tell Maplotlib to make room for the other elements of the graph with autolayout value of our rcParams. Updating the rcParams with autolayout will allow the elements to fit to the graph and appears to trim it down as well

[Top](#Section)

```python
plt.rcParams.update({'figure.autolayout':True}) 

fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
```

## Adding Lables

If we want to add labels to the plot we can use Artist.set() method to set properties to the axes object.

[Top](#Section)

```python
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
#Here we us the set methods to set the xlable to 'Total Revenue', same for the ylabel and the title
```

## Chart Size
We can also change the size of the chart using pyplot.subplots() function with the figsize kwarg

[Top](#Section)

```python
fig, ax = plt.subplots(figsize=(8, 4))
#We still call the fig as our canvase but we are not using figsize inside of the subplot function to set the sise of the canvas
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
```

## Formatting Labels
Here it covers custom formatting using functions. We are going to change the labeling on the x-axis for total revenue.

[Top](#Section)

```python
def currency(x,pos):                  #defining the currncy function with x and pos var input, where x and pos is the value and tick postion
            if x >= 1e6:            #if x is greather than 1million
                s = '${:1.1f}M'.format(x*1e-6)     #then format x by multiply by 1 millionth, and format using format parameter
            else:        #currncy of million, adding "":,"" would add comma seaprator, 1.1f is for the format after the decimal, and add K or M after
                s = '${:1.0f}K'.format(x*1e-3)
            return s
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')


ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
ax.xaxis.set_major_formatter(currency)
```

## Combining Multiple Visualization
Here we are going to draw another line on the same instance of axes.Axes. We call another plot method on the axes object

[Top](#Section)

```python
fig, ax = plt.subplots(figsize=(8, 8))  #call our fix and ax, create a subplot with a figure that is the size of 8x8
ax.barh(group_names, group_data)   # on the axis we are going to draw a bar chart, horizondalally (bar h), with the variables
labels = ax.get_xticklabels()   #going to get the lables with ax.get_xticklables command assign to lables
plt.setp(labels, rotation=45, horizontalalignment='right')  #plt.setp will take the lables and rotate, align right

# Add a vertical line, here we set the style in the function call
ax.axvline(group_mean, ls='--', color='r') #**Here is the SECOND line, second plot method on the same object. 
                                            #Draw line at group mean, using dashed, and color red

# Annotate new companies
for group in [3, 5, 8]:   #Add new name to the following company position at 3, 5, and 8
    ax.text(145000, group, "New Company", fontsize=10,
            verticalalignment="center")

# Now we move our title up since it's getting a little cramped
ax.title.set(y=1.05)

ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
#Here we us the set methods to set the xlable to 'Total Revenue', same for the ylabel and the title

ax.xaxis.set_major_formatter(currency)  #use the currency function to set the format for the x-axis
ax.set_xticks([0, 25e3, 50e3, 75e3, 100e3, 125e3])  #set the tick marks on the x-axis
fig.subplots_adjust(right=.15) #adjust the subplot

plt.show()
```

## Conclusion
In this example we've chosen <b>object orienter interface </b>to do our plotting. We're going to use fig, ax = plt.subplots() as our <b>canvase</b> and use ax as our <b>paint</b> brush to paint our horizontal bar chart and chose the bmh <b>style</b>.

We used <b>ax.get_xticklabels()</b> to get the value and location of our xticklabels so that we can label/format them better. To rotate the xlabels we used the <b>plt.setp()</b> function. In the event our chart dont fit we can call <b>plt.rcParams.update()</b> to help auto fit the chart. Next we added x, y <b>labels</b> as well as a title. Then we changed the chart size using <b>figsize</b>. To make custom formats to the x-axis we used <b>string.format()</b>. Lastly, we drew another line on the same instance of axes.Axes to draw a <b>secondary chart</b>.
