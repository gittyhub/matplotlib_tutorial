## Applying Pseudocolor Schemes

Pseudocolor schemes are particularly useful for apply contrast to an image. This may be helpful inpresentation when contrast is poor.<br>
We can see from the above mpimg.imread that the R,B and G are similar, we can just pick one channel for our data. By doing this we are creating a lumiosity image, that is 2D with no color. More color maps can be found here:
https://matplotlib.org/tutorials/colors/colormaps.html

```python
lum_img = img[:, :,0] #lum_img = img[:, :], what do the different indexes control here?
#All of the RGB channels are similar as you can see from above, img, we select all of the first channel, all of the second channel, and the first position of the thrird channel?
plt.imshow(lum_img)
```

```python
plt.imshow(lum_img, cmap="hot")
#plt.imshow(lum_img, cmap="binary")
#plt.imshow(lum_img, cmap="magma")
```

Color maps can also be set on existing objects using the set_map() method

```python
imgplot = plt.imshow(lum_img)
imgplot.set_cmap('bone')
imgplot.set_cmap('nipy_spectral')
```

## Examining Data Using Histogram

### Extra Contrast
Most of the interesting part of the data can be found around the peak, so we get extra contrast from clipping the region above and/or below the peak. In this example there there is not much info in the high end, not a lot of white in the image. We can adjust the upper limits using clim to achive this.

```python
imgplot = plt.imshow(lum_img, clim=(0.0,175))
```

## Object Oriented Way

```python
fig = plt.figure()
ax = fig.add_subplot(1,2,1)
imgplot = plt.imshow(lum_img)

ax.set_title('Before')
plt.colorbar(ticks=[25, 75, 125, 175], orientation='horizontal')

ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(lum_img)
imgplot.set_clim(0.0,175)

ax.set_title('After')
plt.colorbar(ticks=[25, 75, 125, 175], orientation='horizontal')

#Noticed the difference between the two. Before most of the limits was around 0.5 to 0.6, now that we set the clim to 0.7, most of the image is in the 0.7 range
```

## Array Interpolation Schemes
Interpolation calculates what the color value should be per mathmatical schemes. So when you resize an image, the number of pixel changes but you'd still want to see the original image. In doing so we will be missing pixel which is why when we resize images, we get the pixelation and interpolation attempts to fill in the gap. We will do the following

1) Shrink the image. This will get rid of some pixel to resize it. <br>
2) Plot the image. When plot the image, the smaller image will be blow up to fit in the fig space. <br>
3) Because some of the pixels are gone, the computer has to draw in pixels to fill the space. <br>

```python
from PIL import Image

im = Image.open("/home/hman/Desktop/stinkbug.webp")
width, height = im.size
im.thumbnail((64, 64), Image.ANTIALIAS)  # resizes image in-place
imgplot = plt.imshow(im)
print('Original width: {}\nOriginal height: {}'.format(width, height))
```

Default interpolation is bilinear. There are other interpolation options we can use, "nearest"
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imshow.html#matplotlib.pyplot.imshow

```python
imgplot = plt.imshow(im, interpolation="nearest")
```

```python
imgplot = plt.imshow(im, interpolation="bicubic")
```
